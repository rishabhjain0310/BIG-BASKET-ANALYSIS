# Finding Outlier In Dataset.

```python
sns.scatterplot(y="sale_price",x="market_price",data=df)
plt.show()
```

```python
df[df.sale_price>20000]
```

```python
# Correcting the Outlier.
df.sale_price=np.where(df.sale_price>20000,df.market_price,df.sale_price)
```

```python
# scartterplot after Treating Outlier. 
sns.scatterplot(x=df.market_price,y=df.sale_price)
plt.show()
```

```python
# Measuring discount on a certain item.
df.insert(7,'Discount',df.market_price-df.sale_price)
```

```python
# Find out the Missing Values from the Dataset.
df.isnull().sum()
```

## Dealing With Null Values of a  Dataset

```python
# Filling The Null Value Of Product Coloumn.
df['product']=df['product'].fillna('Unknown')

# Filling The Null Value Of Sale_Price Coloumn.
df.sale_price = df.sale_price.fillna(df.sale_price.median())

# Filling The Null Value Of Brand Coloumn.
df.brand=df.brand.fillna('unknown')

# Filling The Null Value Of Discount Coloumn.
df.Discount = df.Discount.fillna(df.Discount.median())

# Filling The Null Value Of rating Coloumn.
df.rating = df.rating.fillna(df.rating.median())

# Filling The Null Value Of description Coloumn.
df.description=df.description.fillna('no description')
```

```python
# Checking Outlier In the sales price Column.
sns.boxplot(df.sale_price,orient='h')
```

```python
Q1,Q3 = df.sale_price.quantile([0.25,0.75])
```

```python
Q1,Q3
```

```python
IQR = Q3-Q1
```

```python
IQR
```

```python
UL = Q3+1.5*(IQR)
LL = Q1-1.5*(IQR)
```

```python
UL,LL
```

```python
df[df.sale_price>UL]
```

```python
df.sale_price.mean()
```

```python
df.sale_price = np.where(df.sale_price>UL,df.sale_price.mean(),df.sale_price)
```

```python
sns.boxplot(df.sale_price,orient='h')
```

```python
# Checking Outlier In the market price Column.
sns.boxplot(df.market_price,orient='h')
```

```python
Q1,Q3 = df.market_price.quantile([0.25,0.75])
```

```python
Q1,Q3
```

```python
IQR = Q3-Q1
```

```python
IQR
```

```python
UL = Q3+1.5*(IQR)
LL = Q1-1.5*(IQR)
```

```python
UL,LL
```

```python
df.market_price.mean()
```

```python
df.market_price = np.where(df.market_price>UL,df.market_price.mean(),df.market_price)
```
